### Exercise 1 - Webscrabing

Navigate to url via the link and proceed to execute the exercises.

- Find total number of movies playing on that day"

```python
import bs4
import requests


request = requests.get('http://www.kultunaut.dk/perl/searchlist/type-nynaut?Genre=Familiefilm&Genre=Komedie%2FRomantik&Genre=Action%2FDrama&Genre=Science%20Fiction&Genre=Thriller%2FGyser&Genre=Tegnefilm%2FAnimation&Genre=Dokumentar&Genre=Andre%20Film&&Area=&periode=')
request.raise_for_status()
soup = bs4.BeautifulSoup(request.text, 'html.parser')

movies = []

for movie in soup.find_all('b'):
    movies.append(movie.getText())

## First line is not a movie
movies.pop(0)

print("The total number of movies playing today is: " + str(len(movies)))
```

### Exercise 2 - Webscrabing       

- Print all movie titles with their description


```python
movie_titles = soup.find_all('div', attrs={'style': 'clear:both'})

for movie in movie_titles:
    print(movie.getText())
```

- Find the number of all link on the site

```python
links = []

for link in soup.find_all('a'):
    if 'http' in link.get('href'):
        links.append(link)
       
print(len(links))
```

- From this page (click me) use selenium to navigate through the website so that you end up in the same place as the first URL given in exercise 1

```python
from selenium import webdriver
import time

browser = webdriver.Chrome('./chromedriver')

browser.get('http://www.kultunaut.dk/')

accept_cookies = browser.find_element_by_class_name('cookiebanner-close').click()

browser.find_elements_by_xpath("//a[b='Film']")[0].click()
time.sleep(1)
browser.find_elements_by_xpath("//a[b='Vælg alle']")[0].click()
time.sleep(1)
browser.find_elements_by_xpath("//form[@name='search']/table/tbody/tr/td[2]/input[1]")[0].click()
```
